```python
import pandas as pd
import numpy as np

# 1. Create a dictionary that consists of above mentioned values
print("Step 1: Create a dictionary")
employee_data = {
    'Id': [1, 2, 3, 4],
    'Name': ['Pankaj', 'Meghna', 'David', 'Lisa'],
    'Role': ['CEO', np.nan, np.nan, np.nan], # np.nan for empty cells
    'Salary': [100, 200, np.nan, np.nan]     # np.nan for empty cells
}
print("Dictionary created:", employee_data)
print("\n" + "="*50 + "\n")

# 2. Create a dataframe from this dictionary
print("Step 2: Create a DataFrame")
df = pd.DataFrame(employee_data)
print("Initial DataFrame:")
print(df)
print("\n" + "="*50 + "\n")

# 3. Print the initial 2 and last 2 values of this dataframe
print("Step 3: Print initial 2 and last 2 values")
print("Initial 2 rows (df.head(2)):")
print(df.head(2))
print("\nLast 2 rows (df.tail(2)):")
print(df.tail(2))
print("\n" + "="*50 + "\n")

# 4. Check the total number of null values present in the data frame
print("Step 4: Check total number of null values")
total_null_values = df.isnull().sum().sum()
print(f"Total number of null values in the DataFrame: {total_null_values}")
print("\n" + "="*50 + "\n")

# 5. Print the detailed information of the data frame
print("Step 5: Print detailed information (df.info())")
df.info()
print("\n" + "="*50 + "\n")

# 6. Drop all the rows with null values, store this into a new data frame and print it
print("Step 6: Drop rows with null values")
df_rows_dropped = df.dropna(axis=0) # axis=0 or 'index' to drop rows
print("DataFrame after dropping rows with any null values:")
print(df_rows_dropped)
print("\n" + "="*50 + "\n")

# 7. Drop all the columns with null values, store this into a new data frame and print it
print("Step 7: Drop columns with null values")
df_cols_dropped = df.dropna(axis=1) # axis=1 or 'columns' to drop columns
print("DataFrame after dropping columns with any null values:")
print(df_cols_dropped)
print("\n" + "="*50 + "\n")

# 8. Fill null values of the Salary column with the value 300 and print the updated data frame.
print("Step 8: Fill 'Salary' null values with 300")
# Create a copy to avoid modifying the original df for subsequent steps if needed,
# or if df was already modified inplace by previous steps.
df_filled_salary = df.copy()
df_filled_salary['Salary'].fillna(300, inplace=True)
print("DataFrame after filling 'Salary' NaN with 300:")
print(df_filled_salary)
print("\n" + "="*50 + "\n")


# 9. Fill null values of the Role column with the value 'CEO' and print the updated data frame.
print("Step 9: Fill 'Role' null values with 'CEO'")
# Use the original df again, or a fresh copy if you want to apply this independently
# from the salary fill. If you want to apply this to the df_filled_salary, use that one.
df_filled_role = df.copy() # Starting from original df again for clarity
df_filled_role['Role'].fillna('CEO', inplace=True)
print("DataFrame after filling 'Role' NaN with 'CEO':")
print(df_filled_role)
print("\n" + "="*50 + "\n")
```
